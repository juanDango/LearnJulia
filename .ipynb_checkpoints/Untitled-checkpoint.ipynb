# Este codigo es para dibujar las figuras de lisajaus segun el punto 1 de las tareas enviadas donde x=cos(t) y y=cos(at+b)

## Punto a) 

### Para valores de a = 2

#### Para valores de b = 0

```julia
using Plots
valores = range(0,stop=20,length=500)
```

```julia
x = cos.(valores)
```

```julia
y = cos.(2*valores)
```

```julia
plot(x,y)
```

#### Para valores de b = π/4

```julia
y = cos.(2*valores.+pi/4)
```

```julia
plot(x,y)
```

#### Para valores de b = π/2

```julia
y = cos.(2*valores.+pi/2)
```

```julia
plot(x,y)
```

#### Para valores de b = 3π/4

```julia
y = cos.(2*valores.+3*pi/4)
```

```julia
plot(x,y)
```
